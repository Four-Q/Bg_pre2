# Train Test Split

```python
import os
import random
import pandas as pd

RANDOM_SEED = 42
```

```python
# load data
mp_feature_df = pd.read_csv('./Data/featured_data/mp_elements_feat.csv')
exp_feature_df = pd.read_csv('./Data/featured_data/exp_elements_feat.csv')
mp_feature_plus_df = pd.read_csv('./Data/featured_data/mp_elements_plus_feat.csv')
```

```python
print(f'MP测试集大概大小：{mp_feature_df.shape[0]*0.2}')
print(f'EXP测试集大概大小：{exp_feature_df.shape[0]*0.2}')
```

```python
mp_unique_comp = mp_feature_df['composition'].unique()
len(mp_unique_comp)
```

```python
# 从mp_unique_comp中随机采样1104种组成
random.seed(RANDOM_SEED)
random_index = random.sample(range(len(mp_unique_comp)), 1104)
mp_test_comp = mp_unique_comp[random_index]
mp_train_set = mp_feature_df[~mp_feature_df['composition'].isin(mp_test_comp)]
mp_test_set = mp_feature_df[mp_feature_df['composition'].isin(mp_test_comp)]
print(f'MP训练集大小：{len(mp_train_set)}')
print(f'MP测试集大小：{len(mp_test_set)}')
```

```python
mp_plus_train_set = mp_feature_plus_df[~mp_feature_plus_df['composition'].isin(mp_test_comp)]
mp_plus_test_set = mp_feature_plus_df[mp_feature_plus_df['composition'].isin(mp_test_comp)]
print(f'MP Plus训练集大小：{len(mp_plus_train_set)}')
print(f'MP Plus测试集大小：{len(mp_plus_test_set)}')
```

```python
exp_unique_comp = exp_feature_df['composition'].unique()
len(exp_unique_comp)
```

```python
# 从exp_unique_comp中随机采样200种组成
random_index = random.sample(range(len(exp_unique_comp)), 200)
exp_test_comp = exp_unique_comp[random_index]
exp_train_set = exp_feature_df[~exp_feature_df['composition'].isin(exp_test_comp)]
exp_test_set = exp_feature_df[exp_feature_df['composition'].isin(exp_test_comp)]
print(f'EXP训练集大小：{len(exp_train_set)}')
print(f'EXP测试集大小：{len(exp_test_set)}')
```

```python
# 混合测试集
# mp:exp = 1:1
mix_test_set = mp_test_set.sample(n=round(1*exp_test_set.shape[0])).reset_index(drop=True)
```

```python
mix_test_set = pd.concat([mix_test_set, exp_test_set], axis=0).reset_index(drop=True)
mix_test_set = mix_test_set.sample(frac=1, random_state=RANDOM_SEED).reset_index(drop=True)
print(f'混合测试集大小：{len(mix_test_set)}')
```

```python
# 保存
os.makedirs('./Data/train_data', exist_ok=True)
os.makedirs('./Data/test_data', exist_ok=True)
mp_train_set.to_csv('./Data/train_data/mp_train_set.csv', index=False)
mp_test_set.to_csv('./Data/test_data/mp_test_set.csv', index=False)
mp_plus_train_set.to_csv('./Data/train_data/mp_plus_train_set.csv', index=False)
mp_plus_test_set.to_csv('./Data/test_data/mp_plus_test_set.csv', index=False)
exp_train_set.to_csv('./Data/train_data/exp_train_set.csv', index=False)
exp_test_set.to_csv('./Data/test_data/exp_test_set.csv', index=False)
mix_test_set.to_csv('./Data/test_data/mix_test_set.csv', index=False)
```
