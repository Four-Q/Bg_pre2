# Feature Information

```python
import os
import pandas as pd
```

```python
current_path = os.getcwd()
current_path
```

```python
feat_df = pd.read_csv(os.path.join(current_path, './metrics/feature_importance/RFE/feature_importance.csv'))
feat_df.head()
```

```python
# 计算相同特征的所有不同统计量的平均值
feat_df['feature'] = feat_df['Feature'].apply(lambda x: x.split(' ')[1])
feat_df['statistic'] = feat_df['Feature'].apply(lambda x: x.split(' ')[0])
feat_df.drop(columns=['Feature'], inplace=True)
feat_df
```

```python
feat_df['importance'] = feat_df['Importance']
feat_df.drop(columns=['Importance'], inplace=True)
feat_df
```

```python
feat_df.groupby(['feature']).agg({'importance': 'sum'}).reset_index().sort_values(by='importance', ascending=False).reset_index(drop=True).head(20)
```
